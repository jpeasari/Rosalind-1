## Enumerating k-mers Lexicographically
http://rosalind.info/problems/lexf/

```python
seq_A = []
Final_A = []
Final_B = []
Final_C = []
Final_D = []
Final_E = []
Final_F = []
Final_G = []
Final_H = []
Final_I = []
Final_J = []



seq = ["A","B","C","D","E","F","G","H","I","J"]
seq_rest = seq[0:]
for i in range(len(seq_rest)):
    Final_A.append(seq[0] +  seq_rest[i])
    Final_B.append(seq[0+1] + seq_rest[i])    
    Final_C.append(seq[0+2] + seq_rest[i])
    Final_D.append(seq[0+3] + seq_rest[i])
    Final_E.append(seq[0+4] + seq_rest[i])
    Final_F.append(seq[0+5] + seq_rest[i])
    Final_G.append(seq[0+6] + seq_rest[i])
    Final_H.append(seq[0+7] + seq_rest[i])
    Final_I.append(seq[0+8] + seq_rest[i])
    Final_J.append(seq[0+9] + seq_rest[i])
        


print("\n".join(Final_A))
print("\n".join(Final_B))
print("\n".join(Final_C))
print("\n".join(Final_D))
print("\n".join(Final_E))
print("\n".join(Final_F))
print("\n".join(Final_G))
print("\n".join(Final_H))
print("\n".join(Final_I))
print("\n".join(Final_J))


    
    
# print("\n",*Final_A,"\n",*Final_C,"\n",*Final_G,"\n",*Final_T)
    


    
```

```python
from itertools import product

def main(alpha, num):
    orderedalpha = alpha.replace(" ", "")
    permlist = ([x for x in product(orderedalpha, repeat=num)])
    for i in permlist:
        print(''.join(map(str, i)))

# Test
main("A C G T", 3)

```
