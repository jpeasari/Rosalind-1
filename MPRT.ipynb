```python
with open('Uniprot.txt') as f_input, open('uniprot_output.txt', 'w') as f_output:
    block = []

    for line in f_input:
        if line.startswith('>'):
            if block:
                f_output.write(''.join(block) + '\n')
                block = []
            f_output.write(line)
        else:
            block.append(line.strip())

    if block:
        f_output.write(''.join(block) + '\n')
```

```python
## Converting ID and protein sequence to a list of tuples
lst = []
data = open("uniprot_output.txt","r")
file1 = open("uniprot_input.txt","r")
for line in data:
    words = line.split()
    if line[0] == ">":
        x1 = line.replace("|","_")
        y = x1.replace("sp_","")
        z = y.replace(">","")

        words = z.split()
        c = words[0]
    else:
        lst.append((c,line.strip()))
print(lst)
```

```python
## Converting ID and protein sequence to a dictionary
dictionary = {} 
dictionary = dict(lst) 
print(dictionary)

```

```python
def finding_protein_motif(fa): #Pattern of the motif N{P}[ST]{P}
    motif_position_list = []
    try:# Empty list to store found motifs
        for i in range(len(fa)):
            if fa[i] == 'N' and fa[i+1] != 'P' and fa[i+2] == 'S' and fa[i+3] != 'P': # motif N{P}S{P}
                motif_position_list.append(i+1) # Position
            elif fa[i] == 'N' and fa[i+1] != 'P' and fa[i+2] == 'T' and fa[i+3] != 'P': # motif N{P}T{P}
                motif_position_list.append(i+1) # Position
                
        return(motif_position_list)
    
    except IndexError: 
        pass


        
```

```python
motifs = {}
for value, sequence in zip(dictionary.values(), all_ids):
    motif_list = finding_protein_motif(value)
    if motif_list:
        motifs.update( {sequence : motif_list} ) 
    
for key, value in motifs.items(): # print out protein ID and motif positions
    print("\n"+key) 
    for i in value:
        print(i, end=" ")
    
    
```
