```python
number = int(input("Enter a number:"))
factorial = 1
for i in range(1,number + 1):
        factorial = factorial*i
print(factorial) ### Factorial or number of permutations of a given number
```

```python
num = ''
number = int(input("Enter a number:"))
for i in range(number): 
    num = num + str(i+1)
print(type(num))
print(num)
       
```

```python
def permutation(lst): 
  
    if len(lst) == 0: 
        return [] 
  
    elif len(lst) == 1: 
        return [lst] 
    l = [] 
  
    for i in range(len(lst)): 
       m = lst[i] 
       remLst = lst[:i] + lst[i+1:] 
       for p in permutation(remLst): 
           l.append([m] + p) 
    return l 
  
data = list(num) 
for p in permutation(data): 
    print(*p) 
```

```python
from itertools import permutations

seq = permutations(("1","2","3","4","5","6"), 6)

for p in list(seq):
   print(*p)
```
